```python
import csv
from konlpy.tag import Okt

#1. 먼저 hotel_info.csv 파일에서 호텔 설명 부분의 텍스트를 모은다
str = []

with open('hotel_info.csv', 'r', encoding='utf-8') as f:
    rdr = csv.reader(f)

    for line in rdr:
        str.append(line[0])

print(str)

f.close()
```

```python
# 2. 한국어 형태소 처리
okt = Okt()
tokenized_data = [okt.morphs(description) for description in str]

print(tokenized_data)
```

```python
from gensim.models import Word2Vec

# 3. Word2Vec 모델 생성
model = Word2Vec(sentences=tokenized_data, vector_size=100, window=5, min_count=5, workers=4, sg=0)
```

```python
# 4. 유사 단어 검색
search_terms = open('제주_검색어.txt', 'r').read().splitlines()
similar_words = {}

for term in search_terms:
    try:
        similar_words[term] = model.wv.most_similar(term)
    except KeyError:
        similar_words[term] = []

# 결과 출력
for term, similar in similar_words.items():
    print(f"[검색어: {term}]")
    for word, similarity in similar:
        print(f"  {word}: {similarity:.4f}")
    print("\n")
```
